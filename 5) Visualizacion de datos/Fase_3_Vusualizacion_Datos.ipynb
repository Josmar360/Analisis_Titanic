# Importar bibliotecas que se usaran

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Abrir archivo CVS

```python
# Abrir el archivo CSV
data = pd.read_csv("Titanic-Dataset.csv", index_col='PassengerId')
```

# 1.	Crea un histograma de las edad y observa el tipo de distribuci√≥n.

```python
# Configuraci√≥n del estilo de seaborn para gr√°ficos m√°s atractivos
sns.set(style="whitegrid")

# Crear el histograma utilizando seaborn
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=30, kde=True, color='skyblue')
plt.title('Histograma de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

# 2.	Cree un diagrama de dispersi√≥n que muestre la relaci√≥n entre la tarifa pagada y la supervivencia.

```python
# Crear un diagrama de dispersi√≥n
plt.figure(figsize=(10, 6))
plt.scatter(data['Fare'], data['Survived'], alpha=0.5)

# Personalizar el gr√°fico
plt.title('Relaci√≥n entre Tarifa Pagada y Supervivencia')
plt.xlabel('Tarifa Pagada')
plt.ylabel('Supervivencia')
plt.grid(True)

# Mostrar el gr√°fico
plt.show()
```

# 3.	Cree un grafico que muestre la distribuci√≥n de genero por clases.

```python
df = pd.DataFrame(data)

# Crear un gr√°fico de barras usando seaborn
plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', hue='Sex_female', data=df, palette='Set1')

# Personalizar el gr√°fico
plt.title('Distribuci√≥n de g√©nero por clases')
plt.xlabel('Clase')
plt.ylabel('N√∫mero de pasajeros')
plt.legend(title='G√©nero', labels=['Male', 'Female'])

# Mostrar el gr√°fico
plt.show()
```

# 4.	Identificar aspectos interesantes de los datos y analizar qu√© tipo de gr√°fico ser√° el m√°s apropiado.

```python
# Crear un gr√°fico de barras
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))

# Usar seaborn para crear un gr√°fico de barras agrupado por g√©nero y supervivencia
sns.countplot(x="Survived", hue="Sex_female", data=data)

# Personalizar el gr√°fico
plt.title('Comparaci√≥n entre G√©nero y Supervivencia')
plt.xlabel('Supervivencia (0 = No, 1 = S√≠)')
plt.ylabel('Cantidad de Pasajeros')
plt.legend(title='G√©nero', labels=['Masculino', 'Femenino'])

# Mostrar el gr√°fico
plt.show()
```
